#แทนที่ทีละ column

```python
#Beauty
def getcatBeauty(text,lookup= Beauty):
  for cat in lookup :
    for keyword in lookup[str(cat)]:
      if keyword in str(text):
        return {"catagory" :cat}
  return {"catagory" :""}
```

```python
#Fashion
def getcatFashion(text,lookup= Fashion):
  for cat in lookup :
    for keyword in lookup[str(cat)]:
      if keyword in str(text):
        return {"catagory" :cat}
  return {"catagory" :""}
```

```python
def catagoryofBeauty(df):
  #ตรงนี้จะเป็นของColumn VID_LP[{_11001}].BLVID
  df['TUBA[{_1}]._scale_code'] = df['TUBA[{_1}]._scale'].apply(lambda text:getcatBeauty(text)["catagory"])
  df['TUBA[{_2}]._scale_code'] = df['TUBA[{_2}]._scale'].apply(lambda text:getcatBeauty(text)["catagory"])
  df['TUBA[{_3}]._scale_code'] = df['TUBA[{_3}]._scale'].apply(lambda text:getcatBeauty(text)["catagory"])
  df['TUBA[{_4}]._scale_code'] = df['TUBA[{_4}]._scale'].apply(lambda text:getcatBeauty(text)["catagory"])
  df['TUBA[{_5}]._scale_code'] = df['TUBA[{_5}]._scale'].apply(lambda text:getcatBeauty(text)["catagory"])
  df['TUBA[{_6}]._scale_code'] = df['TUBA[{_6}]._scale'].apply(lambda text:getcatBeauty(text)["catagory"])
  df['TUBA[{_7}]._scale_code'] = df['TUBA[{_7}]._scale'].apply(lambda text:getcatBeauty(text)["catagory"])
  df['TUBA[{_8}]._scale_code'] = df['TUBA[{_8}]._scale'].apply(lambda text:getcatBeauty(text)["catagory"])
  df['TUBA[{_9}]._scale_code'] = df['TUBA[{_9}]._scale'].apply(lambda text:getcatBeauty(text)["catagory"])
  df['TUBA[{_10}]._scale_code'] = df['TUBA[{_10}]._scale'].apply(lambda text:getcatBeauty(text)["catagory"])

  #ตรงนี้จะเป็นของColumn VID_LP[{_11001}].BLVID
  df['VID_LP[{_11001}].BLVID_code'] = df['VID_LP[{_11001}].BLVID'].apply(lambda text:getcatBeauty(text)["catagory"])
  df['VID_LP[{_11002}].BLVID_code'] = df['VID_LP[{_11002}].BLVID'].apply(lambda text:getcatBeauty(text)["catagory"])
  df['VID_LP[{_11003}].BLVID_code'] = df['VID_LP[{_11003}].BLVID'].apply(lambda text:getcatBeauty(text)["catagory"])
  df['VID_LP[{_12001}].BLVID_code'] = df['VID_LP[{_12001}].BLVID'].apply(lambda text:getcatBeauty(text)["catagory"])
  df['VID_LP[{_12002}].BLVID_code'] = df['VID_LP[{_12002}].BLVID'].apply(lambda text:getcatBeauty(text)["catagory"])
  return df
```
